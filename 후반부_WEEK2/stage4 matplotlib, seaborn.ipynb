# 데이터 시각화 (matplotlib, seaborn)  

## matplotlib

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

### 꺾은선그래프

```python
# (1, 90), (2, 100), (3, 50)
plt.plot([1, 2, 3], [90, 100, 50])
plt.show()
```

```python
plt.plot(['Math', 'Korean', 'English'], [90, 100, 50])
plt.plot(['Math', 'Korean', 'English'], [70, 60, 90])

plt.xlabel('Subjects')
plt.ylabel('Score')
plt.title('Midterm')

plt.legend(['Choi', 'Kim'])

plt.show()
```

### 막대그래프

```python
x = ['Math', 'Korean', 'English']
y = [90, 100, 50]

plt.xlabel('Subjects')
plt.ylabel('Score')

plt.bar(x, y, width = 0.5, color= '#511499')  # 두께, 색깔 설정
plt.show()
```

```python
plt.plot([1, 2, 3], [90, 100, 50], marker = 'v', markersize = 10, linestyle = ':')  # 마커 모양 o, v, s, + / 마커 크기/ 선 스타일 :, --

plt.grid(True) 
plt.axis([0, 3, 80, 150])  # axis([x축범위, y축범위])

plt.show()
```

## seaborn

### 판다스로 csv 데이터 불러오기

```python
import pandas as pd
```

```python
train = pd.read_csv('data/train.csv')
train  # 전체 데이터
train.head(6)  # 상단부터 해당 개수만큼만 출력
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

### '나이'와 '생존' 요소를 기준으로 산포도 그리기

```python
sns.relplot(data = train, x = 'Age', y = 'Survived')
```

```python
sns.relplot(data = train, x = 'Age', y = 'Survived', hue = 'Sex', aspect= 3)   # 가로가 세로의 3배/ hue : 성별로 구분
```

### Categorical Scatter Plot 그려보기

```python
sns.catplot(data = train, x = 'Pclass', y = 'Fare', aspect = 4, jitter = True, hue = 'Sex')   # jitter= False : 한곳에 모임
```

### Pair plot 기능 사용하기


```python
sns.pairplot(data = train, hue = 'Sex', 
            x_vars = ['Pclass', 'Age', 'Fare'],
            y_vars = ['Survived', 'Age', 'Fare'], height = 3)
```

### 박스 그림 상자(Box Plot) 그려보기

box plot: 데이터의 유효범위를 판단하는데 유용한 차트

```python
sns.catplot(data = train, x = 'Pclass', y = 'Fare',hue = 'Sex', kind = 'box')
```
