```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
train = pd.read_csv('data/train.csv')
train.head(10)
```

```python
Master_survived = train.loc[(train['Name'].str.contains('Master')) & (train['Survived'] == 1), 'PassengerId'].count()
Master_dead = train.loc[(train['Name'].str.contains('Master')) & (train['Survived'] == 0), 'PassengerId'].count()

Mr_survived = train.loc[(train['Name'].str.contains('Mr')) & (train['Survived'] == 1), 'PassengerId'].count()
Mr_dead = train.loc[(train['Name'].str.contains('Mr')) & (train['Survived'] == 0), 'PassengerId'].count()

Miss_survived = train.loc[(train['Name'].str.contains('Miss')) & (train['Survived'] == 1), 'PassengerId'].count()
Miss_dead = train.loc[(train['Name'].str.contains('Miss')) & (train['Survived'] == 0), 'PassengerId'].count()

Mrs_survived = train.loc[(train['Name'].str.contains('Mrs')) & (train['Survived'] == 1), 'PassengerId'].count()
Mrs_dead = train.loc[(train['Name'].str.contains('Mrs')) & (train['Survived'] == 0), 'PassengerId'].count()
```

```python
df = pd.DataFrame({
    'Master': [Master_dead, Master_survived],
    'Mr': [Mr_dead, Mr_survived, ],
    'Miss': [Miss_dead, Miss_survived],
    'Mrs': [Mrs_dead, Mrs_survived],
})
df
```

```python
df.plot.bar()
```
