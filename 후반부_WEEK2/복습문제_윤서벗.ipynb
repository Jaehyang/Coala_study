```python
import pandas as pd
train = pd.read_csv('data/train.csv')
train.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
pclass_1 = train.loc[(train['Pclass'] == 1) & (train['Survived'] == 1), 'PassengerId'].count()
pclass_2 = train.loc[(train['Pclass'] == 2) & (train['Survived'] == 1), 'PassengerId'].count()
pclass_3 = train.loc[(train['Pclass'] == 3) & (train['Survived'] == 1), 'PassengerId'].count()
print(pclass_1, pclass_2, pclass_3)
```

```python
plt.bar(['1', '2', '3'], [pclass_1, pclass_2, pclass_3], width = 0.5, color = 'black')

plt.xlabel('Pclass')
plt.ylabel('Survived people')

plt.show()
```
