```python
# 생존자
survived_0 = train.loc[(train['SibSp'] == 0) & (train['Survived'] == 1), 'PassengerId'].count()
survived_1 = train.loc[(train['SibSp'] == 1) & (train['Survived'] == 1), 'PassengerId'].count()
survived_2 = train.loc[(train['SibSp'] == 2) & (train['Survived'] == 1), 'PassengerId'].count()
survived_3 = train.loc[(train['SibSp'] == 3) & (train['Survived'] == 1), 'PassengerId'].count()
survived_4 = train.loc[(train['SibSp'] == 4) & (train['Survived'] == 1), 'PassengerId'].count()
survived_5 = train.loc[(train['SibSp'] == 5) & (train['Survived'] == 1), 'PassengerId'].count()
survived_6 = train.loc[(train['SibSp'] == 6) & (train['Survived'] == 1), 'PassengerId'].count()
survived_7 = train.loc[(train['SibSp'] == 7) & (train['Survived'] == 1), 'PassengerId'].count()
survived_8 = train.loc[(train['SibSp'] == 8) & (train['Survived'] == 1), 'PassengerId'].count()

# 사망자

dead_0 = train.loc[(train['SibSp'] == 0) & (train['Survived'] == 0), 'PassengerId'].count()
dead_1 = train.loc[(train['SibSp'] == 1) & (train['Survived'] == 0), 'PassengerId'].count()
dead_2 = train.loc[(train['SibSp'] == 2) & (train['Survived'] == 0), 'PassengerId'].count()
dead_3 = train.loc[(train['SibSp'] == 3) & (train['Survived'] == 0), 'PassengerId'].count()
dead_4 = train.loc[(train['SibSp'] == 4) & (train['Survived'] == 0), 'PassengerId'].count()
dead_5 = train.loc[(train['SibSp'] == 5) & (train['Survived'] == 0), 'PassengerId'].count()
dead_6 = train.loc[(train['SibSp'] == 6) & (train['Survived'] == 0), 'PassengerId'].count()
dead_7 = train.loc[(train['SibSp'] == 7) & (train['Survived'] == 0), 'PassengerId'].count()
dead_8 = train.loc[(train['SibSp'] == 8) & (train['Survived'] == 0), 'PassengerId'].count()

plt.bar(['Survived', 'Dead'], [survived_0, dead_0], width = 0.3)
plt.bar(['Survived', 'Dead'], [survived_1, dead_1], width = 0.3)
plt.bar(['Survived', 'Dead'], [survived_2, dead_2], width = 0.3)
plt.bar(['Survived', 'Dead'], [survived_3, dead_3], width = 0.3)
plt.bar(['Survived', 'Dead'], [survived_4, dead_4], width = 0.3)
plt.bar(['Survived', 'Dead'], [survived_5, dead_5], width = 0.3)
plt.bar(['Survived', 'Dead'], [survived_6, dead_6], width = 0.3)
plt.bar(['Survived', 'Dead'], [survived_7, dead_7], width = 0.3)
plt.bar(['Survived', 'Dead'], [survived_8, dead_8], width = 0.3)

plt.show()
```
