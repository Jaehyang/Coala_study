```python
import pandas as pd
train = pd.read_csv('data/train.csv')

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

train.head(10)
```

```python
survived_0 = train.loc[(train['Age'] <= 10) & (train['Survived'] == 1), 'PassengerId'].count()
survived_10 = train.loc[(train['Age'] > 10) & (train['Age'] <= 20) & (train['Survived'] == 1), 'PassengerId'].count()
survived_20 = train.loc[(train['Age'] > 20) & (train['Age'] <= 30) & (train['Survived'] == 1), 'PassengerId'].count()
survived_30 = train.loc[(train['Age'] > 30) & (train['Age'] <= 40) & (train['Survived'] == 1), 'PassengerId'].count()
survived_40 = train.loc[(train['Age'] > 40) & (train['Age'] <= 50) & (train['Survived'] == 1), 'PassengerId'].count()
survived_50 = train.loc[(train['Age'] > 50) & (train['Survived'] == 1), 'PassengerId'].count()

dead_0 = train.loc[(train['Age'] <= 10) & (train['Survived'] == 0), 'PassengerId'].count()
dead_10 = train.loc[(train['Age'] > 10) & (train['Age'] <= 20) & (train['Survived'] == 1), 'PassengerId'].count()
dead_20 = train.loc[(train['Age'] > 20) & (train['Age'] <= 30) & (train['Survived'] == 1), 'PassengerId'].count()
dead_30 = train.loc[(train['Age'] > 30) & (train['Age'] <= 40) & (train['Survived'] == 1), 'PassengerId'].count()
dead_40 = train.loc[(train['Age'] > 40) & (train['Age'] <= 50) & (train['Survived'] == 1), 'PassengerId'].count()
dead_50 = train.loc[(train['Age'] > 50) & (train['Survived'] == 1), 'PassengerId'].count()
```

```python
df = pd.DataFrame({
    '0-10': [survived_0,dead_0],
    '10-20': [survived_10, dead_10],
    '20-30': [survived_20, dead_20],
    '30-40': [survived_30, dead_30],
    '40-50': [survived_40, dead_40],
    '50-': [survived_50,dead_50]
})
df
```

```python
df.plot.bar()
```
