#  Iris 문제 Feature Engineering -> Iris 종 구분하기  
아이리스의 꽃잎과 꽃받침의 너비, 길이 정보를 바탕으로 setosa종인지 versicolor종인지 혹은 virginica종인지 구분하는 문제  
  
- SepalLengthCm: 꽃받침의 길이  
- SepalWidthCm: 꽃받침의 너비  
- PetalLengthCm: 꽃잎의 길이  
- PetalWidthCm: 꽃입의 너비  

## Iris 데이터 불러오기 (함께실습1)

```python
import pandas as pd
```

```python
df = pd.read_csv('data/iris.csv')
df.head(10)
```

### c.f. 요약 정보 : info(), decribe()

```python
df.info()
```

```python
df.describe()
```

## 꽃받침 길이, 너비에 따른 종 분포도 보기 (함께실습2)


```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
sns.catplot(data=df, x = 'SepalLengthCm', y = 'SepalWidthCm', hue = 'Species', aspect = 2)
# 꽃받침의 길이와 너비 잘 분석하면 setosa는 분리가 가능할듯!
```

## 꽃잎의 길이, 너비에 따른 종 분포도 보기(미션1)


```python
sns.catplot(data=df, x = 'PetalLengthCm', y = 'PetalWidthCm', hue = 'Species', aspect = 2)
```

 ## kdeplot으로 꽃받침 길이별 밀집 수준 파악 (함께실습3)


```python
facet = sns.FacetGrid(df, hue = 'Species', aspect = 3)
facet.map(sns.kdeplot, 'SepalLengthCm')
facet.add_legend()
plt.show()
# 꽃받침 길이가 작을 수록 setosa일 확률이 높음. 꽃받침 길이가 길 수록 virginica일 확률이 높은 걸 알 수 있음. 중간엔 versicolor.
```

## box plot으로 종별 꽃받침 길이 범위 파악 (함께실습4)


```python
sns.catplot(kind = 'box', data = df, x = 'Species', y = 'SepalLengthCm', aspect = 2)
# setosa는 5.0, versicolor은 6.0, virginica는 6.5 정도에 몰려 있는 걸 알 수 있음. 
```

## violin plot - x: 종, y: 꽃받침 길이 (함께실습5)

### c.f. violin plot: 박스플롯과 kdeplot의 장점을 합친 플롯


```python
sns.catplot(kind = 'violin', data = df, x = 'Species', y = 'SepalLengthCm', aspect = 2)
```

## violin plot - x: 종, y: 꽃받침 너비 (함께실습6)


```python
sns.catplot(kind = 'violin', data = df, x = 'Species', y = 'SepalWidthCm', aspect = 2)
```

```python
sns.catplot(kind = 'violin', data = df, x = 'Species', y = 'PetalLengthCm', aspect = 2)
```
