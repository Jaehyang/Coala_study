# 윤서벗
Iris.csv 데이터로 트리를 2 개를 만들어 보세요.  
  
1. 원하는 범위의 숫자데이터로 바꾸어 주기(PetalWidCm, PetalLengthCm)
 - petalWidCm 은 0.0<= x < 1.0 이면 0.5 , 1.0<= x < 2.0 이면 1.5 , 2.0<= x < 3.0 이면 2.5
 - petalLengCm 은 1.0<= x < 3.0 이면 2.0 , 3.0<= x < 5.0 이면 4.0 , 5.0<= x < 7.0 이면 6.0
 - Species: [ setosa=> 1 ] / [ virginica => 2 ] / [ versicolor => 3 ]
으로 범주화 하기
2. input data 는 PetalWidCm, PetalLengthCm 로 / target data 는 Species 로 정하고
3. 0-40 까지의 데이터를 validation data 로 41-끝까지의 데이터를 test set 으로 정함
4. 위 데이터로 트리 2 개를 만들어서 정확도를 비교해보세요!

```python
import pandas as pd
df = pd.read_csv('data/iris.csv')
df
```

```python
# df_test.loc[(df_test['Age'] >= 10) & (df_test['Age'] < 20), 'Age'] = 1
i = 0
for i in range(1, 4):
    df.loc[(df['PetalLengthCm'] >= 2*i-1) & (df['PetalLengthCm'] < 2*i+1), 'PetalLengthCm'] = 2*i
    df.loc[(df['PetalWidthCm'] >= i-1) & (df['PetalWidthCm'] < i), 'PetalWidthCm'] = i - 0.5
df
```
