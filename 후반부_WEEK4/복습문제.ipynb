```python
import pandas as pd
```

```python
df = pd.read_csv('data/train.csv')
df_test = pd.read_csv('data/test.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df_test.isnull().sum()
```

```python
df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
df_test['Fare'] = df_test['Fare'].fillna(df_test['Fare'].mean())
```

```python
df.loc[df['Fare'] < 10, 'Fare'] = 0
df.loc[(df['Fare'] >= 10) & (df['Fare'] < 20), 'Fare'] = 1
df.loc[(df['Fare'] >= 20) & (df['Fare'] < 30), 'Fare'] = 2
df.loc[(df['Fare'] >= 30) & (df['Fare'] < 50), 'Fare'] = 3
df.loc[(df['Fare'] >= 50) & (df['Fare'] < 70), 'Fare'] = 4
df.loc[df['Fare'] >= 70, 'Fare'] = 5
```

```python
df_test.loc[df_test['Fare'] < 10, 'Fare'] = 0
df_test.loc[(df_test['Fare'] >= 10) & (df_test['Fare'] < 20), 'Fare'] = 1
df_test.loc[(df_test['Fare'] >= 20) & (df_test['Fare'] < 30), 'Fare'] = 2
df_test.loc[(df_test['Fare'] >= 30) & (df_test['Fare'] < 50), 'Fare'] = 3
df_test.loc[(df_test['Fare'] >= 50) & (df_test['Fare'] < 70), 'Fare'] = 4
df_test.loc[df_test['Fare'] >= 70, 'Fare'] = 5
```

```python
df.isnull().sum()
```

```python
df.loc[df['Sex'] == 'male', 'Sex'] = 0
df.loc[df['Sex'] == 'female', 'Sex'] = 1

df_test.loc[df_test['Sex'] == 'male', 'Sex'] = 0
df_test.loc[df_test['Sex'] == 'female', 'Sex'] = 1

df.loc[df['Embarked'] == 'S', 'Embarked'] = 0
df.loc[df['Embarked'] == 'C', 'Embarked'] = 1
df.loc[df['Embarked'] == 'Q', 'Embarked'] = 2

df_test.loc[df_test['Embarked'] == 'S', 'Embarked'] = 0
df_test.loc[df_test['Embarked'] == 'C', 'Embarked'] = 1
df_test.loc[df_test['Embarked'] == 'Q', 'Embarked'] = 2
```

```python
df['Embarked'] = df['Embarked'].fillna(0)
df_test['Embarked'] = df_test['Embarked'].fillna(0)
```

```python
df
```

```python
df.isnull().sum()
```

```python
x_train = df[['Sex', 'Fare', 'Embarked']]
x_test = df_test[['Sex', 'Fare', 'Embarked']]
y_train = df['Survived']
```

```python
# DecisionTreeClassifier로 학습, 평가
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier() # 초기화
tree.fit(x_train, y_train) # 학습

tree.score(x_train, y_train) # 성능 확인
```

```python
prediction = tree.predict(x_test)
prediction
```
